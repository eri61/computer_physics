```python
# initial param
v0 = 100.0 / 3.6  # [m/s]
m = 0.1  # [kg]
b = 0.1  # [kg/s]
```

```python
def max_length(v0=100.0 / 3.6, m=0.1, b=0.1, n_t=101):
    # MKS units: m, kg, s
    g = 9.8  # [m/s^2]

    t_start = 0  # [s]
    t_end = 5.0  # [s]

    # Results will be stored into a list as namedtuple
    results = []
    Result = namedtuple('Result', ['key', 'xt', 'yt'])
    num_theta = (90 - 5) // 5
    L = np.zeros((num_theta, 3))

    for i, theta_degree in enumerate(range(5, 90, 5)):
        theta = theta_degree * (np.pi / 180)  # radian

        # X0 = [x0, v0_x, y0, v0_y]
        X0 = np.array([0, v0 * np.cos(theta), 0, v0 * np.sin(theta)])

        # Solve Newton equation
        # x(t), y(t)
        xt, _, yt, _ = solve_newton(eq_params=(g, b/m), X0=X0, t_range=(t_start, t_end), n_t=n_t)
        l, err = reaching_point(xt, yt, num_exclude_point=n_t//20)

        # Store result
        results.append(Result(str(theta_degree), xt, yt))
        L[i, 0] = theta
        L[i, 1] = l
        L[i, 2] = err
    
    max_L = L[:, 1].max()
    print(f"Lが最大となる角度:{L[:, 0][L[:, 1] == max_L].round(2)},  最大のL:{max_L.round(2)} err:{L[:, 2][L[:, 1] == max_L].round(5)}")
    plot(results)

max_length(v0, m, b, n_t=10000)
```

```python
def depend_theta_bm(v0=100.0 / 3.6, bm=1., n_t=101):
    # MKS units: m, kg, s
    g = 9.8  # [m/s^2]

    t_start = 0  # [s]
    t_end = 10.0  # [s]

    # Results will be stored into a list as namedtuple
    results = []
    Result = namedtuple('Result', ['key', 'xt', 'yt'])
    num_theta = (90 - 5) // 5
    L = np.zeros((num_theta, 3))

    for i, theta_degree in enumerate(range(5, 90, 5)):
        theta = theta_degree * (np.pi / 180)  # radian

        # X0 = [x0, v0_x, y0, v0_y]
        X0 = np.array([0, v0 * np.cos(theta), 0, v0 * np.sin(theta)])

        # Solve Newton equation
        # x(t), y(t)
        xt, _, yt, _ = solve_newton(
            eq_params=(g, bm), X0=X0, t_range=(t_start, t_end), n_t=n_t 
            )

        l, err = reaching_point(xt, yt)
        L[i, 0] = theta
        L[i, 1] = l
        L[i, 2] = err

        # Store result
        results.append(Result(str(theta_degree), xt, yt))
    
    max_L = L[:, 1].max()
    max_point = L[L[:, 1] == max_L]
    # print(f"Lが最大となる角度:{max_L[0, 0].round(2)},  最大のL:{max_L[0, 1].round(2)}")

    return max_point
```

```python
n = 100
_depend_theta_bm = np.zeros((n, 3))
vec_bm = np.zeros(n)
set = np.linspace(0.001, 0.1, n)
for i, bm in enumerate(tqdm(set)):
    max_point = depend_theta_bm(bm=bm, n_t=1000)
    _depend_theta_bm[i, :] = max_point
    vec_bm[i] = bm
    
```
